# 5)Check and handle outliers in at least 3 columns in the dataset


```python
data1=titanic['Fare']
```

```python
z_scores=(data1-data1.mean())/data1.std()
```

```python
z_scores.mean()
```

```python
threshold=6
```

```python
outliers=titanic[abs(z_scores)>threshold]
```

```python
print('Outliers are:-')
outliers
```

```python
data2=titanic['Age']
```

```python
Q1=data2.quantile(0.25)
Q3=data2.quantile(0.75)
IQR=Q3-Q1
```

```python
threshold=2.5
```

```python
outliers=titanic[(data2<Q1-threshold*IQR) | (data2>Q3+threshold*IQR)]
print('Outliers are:-')
outliers
```
