# Best Linear relation between price and year

```python
plt.figure(figsize=(8, 6)) 

sns.lineplot(x='Year', y='Price', data=grand_i10)

plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Year vs Price for Grand i10')

plt.grid(True)  

plt.show() 
```

#   Feature variables (X) and target variable (y).

```python
X=grand_i10.drop(columns=['Price'])
y=grand_i10['Price']
```

```python
print(X.shape)
print(y.shape)
```

```python
X=np.array(X)
y=np.array(y)
```

```python
y=y.reshape(-1,1)
```

```python
y.shape
```

# Split data into training sets and testing sets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=40)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

#  create and train a linear regression model 

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
print(y_predict)
```

```python
data=pd.DataFrame({'Year':2022},index=[0])
```

```python
model.predict(data)
```

# Prediction using Lasso regression

```python
from sklearn.linear_model import Lasso
model1=Lasso()
model1.fit(X_train,y_train)
```

```python
y_predict=model1.predict(X_test)
print(y_predict)
```

```python
data1=pd.DataFrame({'Year':2022},index=[0])
model1.predict(data1)
```

# Finding accuracy of Linear Regression

```python
accuracy_linear=model.score(X_test,y_test)
```

```python
accuracy_linear
```

# Finding accuracy of Lasso regression

```python
accuracy_lasso=model1.score(X_test,y_test)
```
